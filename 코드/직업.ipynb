```python
import pandas as pd

# Load the CSV file
file_path = "../Final_라벨결측제거_원핫인코딩.csv"
df = pd.read_csv(file_path)

# Create a new column '직업'
df['직업'] = df['그룹'].apply(lambda x: '배우' if pd.isna(x) else '아이돌')
```

```python
sum(df['직업'].isna())
```

```python
df.to_csv("Final_라벨결측제거_원핫인코딩_직업.csv")
```

```python
# One-hot encoding for the '직업' column
직업_encoded = pd.get_dummies(df['직업'], prefix='직업')

# Concatenate the new one-hot encoded columns to the original dataframe
df_encoded = pd.concat([df, 직업_encoded], axis=1)
```

```python
df_encoded.to_csv("Final_라벨결측제거_원핫인코딩_직업.csv", encoding='utf-8-sig')
```
