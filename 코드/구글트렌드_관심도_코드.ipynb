## 3. 기준 연예인 설정

### <수작업>

직접 검색해보면서 수치 비교함. 김수현이 기준 연예인임.

## 4. 구글 트렌드 요청 그룹

```python
import pandas as pd
import random

# 기준 연예인 (전체 이름)
reference_celeb = "김수현"

# 기준 연예인 제외하고 셔플
comparison_pool = [celeb for celeb in all_celebrities if celeb != reference_celeb]
random.shuffle(comparison_pool)

# 그룹 크기: 기준 제외 4명씩
group_size = 4
comparison_groups = [comparison_pool[i:i+group_size] for i in range(0, len(comparison_pool), group_size)]

# 기준 연예인 포함한 그룹 생성
request_groups = []
for group in comparison_groups:
    group_with_reference = [reference_celeb] + group
    request_groups.append(group_with_reference)

# 결과를 DataFrame으로 변환
max_len = max(len(g) for g in request_groups)
columns = [f"검색어{i+1}" for i in range(max_len)]
df = pd.DataFrame(request_groups, columns=columns)

df
```
